```python
import numpy as np
import matplotlib.pyplot as plt

# 가상의 방사능 농도 데이터 (현재 농도, 방류 횟수에 따른 증가량)
current_radiation_levels = np.array([10, 15, 20, 5, 8])
increase_per_discharge = np.array([1.5, 2.0, 2.5, 1.0, 1.2])

# 방류 횟수에 따른 방사능 농도 변화 시뮬레이션
discharge_count = np.arange(0, 21, 1)
future_radiation_levels = np.zeros((len(current_radiation_levels), len(discharge_count)))

for i, (current_level, increase) in enumerate(zip(current_radiation_levels, increase_per_discharge)):
    future_radiation_levels[i] = current_level + increase * discharge_count

# 그래프 출력
plt.figure(figsize=(10, 6))
for i, levels in enumerate(future_radiation_levels):
    plt.plot(discharge_count, levels, label=f'Point {i+1}')
plt.axhline(y=50, color='r', linestyle='--', label='Safety Threshold')
plt.xlabel('Number of Discharges')
plt.ylabel('Radiation Level (Bq/kg)')
plt.title('Projected Radiation Levels at Different Points')
plt.legend()
plt.show()
```
