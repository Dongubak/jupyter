```python
import pandas as pd

row_df = pd.read_csv('owid-covid-data.csv')

selected_columns = [
    'iso_code', 'location', 'date', 'total_cases', 'population'
]

revised_df = row_df[selected_columns]

kor_df = revised_df[revised_df.location == 'South Korea']
nor_df = revised_df[revised_df.location == 'North Korea']

kor_date_index_df = kor_df.set_index('date')
nor_date_index_df = nor_df.set_index('date')


franch_df = revised_df[revised_df.location == 'France']

franch_date_index_df = franch_df.set_index('date')

kor_total_cases = kor_date_index_df['total_cases']
franch_total_cases = franch_date_index_df['total_cases']
nor_total_cases = nor_date_index_df['total_cases']

kor_population = kor_date_index_df['population']['2020-01-22']
franch_population = franch_date_index_df['population']['2020-01-22']
nkor_population = nor_date_index_df['population']['2020-01-22']

rate_f = round(franch_population / (kor_population + nkor_population + franch_population))
rate_nk = round(nkor_population / (kor_population + nkor_population + franch_population))
rate_k = round(kor_population / (kor_population + nkor_population + franch_population))

df = pd.DataFrame({
    'KOR' : kor_total_cases,
    'FRA' : franch_total_cases,
    'NKOR' : nor_total_cases
}, index = kor_date_index_df.index)

df.plot.line(rot = 45)

# revised_df['location'].unique()
```

```python
import pandas as pd

raw_df = pd.read_csv('owid-covid-data.csv')
selected_column = ['iso_code', 'location', 'date', 'total_cases', 'population']

revised_df = raw_df[selected_column]

kor_df = revised_df[revised_df.location == 'South Korea']
kor_df_index_date = kor_df.set_index('date')
kor_total_cases = kor_df_index_date['total_cases']

us_df = revised_df[revised_df.location == 'United States']
us_df_index_date = us_df.set_index('date')
us_total_cases = us_df_index_date['total_cases']

kor_population = kor_df_index_date['population']['2020-01-22']
us_population = us_df_index_date['population']['2020-01-22']

rate = us_population / kor_population

final_df = pd.DataFrame({
   'KOR': kor_total_cases * rate,
   'USA': us_total_cases
}, index=kor_df_index_date.index)

final_df['2022-01-01':].plot.line(rot=45)
```
